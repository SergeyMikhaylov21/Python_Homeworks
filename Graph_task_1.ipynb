```python
import os
from pymystem3 import Mystem
import networkx as nx
#%matplotlib inline
import matplotlib.pyplot as plt
```

```python
MyS = Mystem()

path_to_txts = 'C:\\Users\Сергей\\YandexDisk\\SovSport\\plaintext\\2016\\январь'
direc = os.listdir(path_to_txts)

stop_words, all_words = [], []

with open('StopWords_Russian.txt', encoding='utf-8') as stop_wrds:
    lns = stop_wrds.readlines()
    for line in lns:
        ln = line.strip().replace('\ufeff', '')
        stop_words.append(ln)
print(stop_words)
```

```python
for file in direc:
    with open(path_to_txts + '\\'+ str(file), encoding='utf-8') as opd_file:
        lines = opd_file.readlines()
        for line in lines:
            wrds = line.strip().split()
            for wrd in wrds:
                lem_wrd = MyS.lemmatize(wrd)
                for i in lem_wrd:
                    if i not in stop_words and i != '\n' and len(i) > 2:
                        all_words.append(i)

print(all_words)
```

```python
art_words = []
trash = ['❤️\n', '»,\n', '...', '!!!', '),\n', '---\n', '»),\n', '»/\n', 'amp', 'quot', 
         '»:\n', '):\n', '***\n', '✨✨✨✨', '--\n', '»)\n', '---', '))\n']
for ch_wrd in all_words:
    if ch_wrd not in trash:
        art_words.append(ch_wrd)

print(art_words)
print('Количество всех извлеченных слов: ' + str(len(art_words)))
```

```python
G = nx.Graph()

for i in range(len(art_words)):
    if art_words[i] == art_words[0]:
        G.add_nodes_from([art_words[0], art_words[1], art_words[2]])
        G.add_edges_from([(art_words[0],art_words[1]),(art_words[0],art_words[2]),(art_words[1],art_words[2])])
    elif art_words[i] == art_words[1]:
        if art_words[3] in G.nodes():
            G.add_edges_from([(art_words[0],art_words[3]),(art_words[1],art_words[3]),(art_words[2],art_words[3])])
        else:
            G.add_node(art_words[3])
            G.add_edges_from([(art_words[0],art_words[3]),(art_words[1],art_words[3]),(art_words[2],art_words[3])])

    elif art_words[i] == art_words[len(art_words)-2]:
        G.add_edges_from([(art_words[i],art_words[i-1]),(art_words[i],art_words[i-2]),(art_words[i-1],art_words[i-2]), (art_words[i],art_words[i+1])])
    elif art_words[i] == art_words[len(art_words)-1]:
        G.add_edges_from([(art_words[i],art_words[i-1]),(art_words[i],art_words[i-2]),(art_words[i-1],art_words[i-2])])

    else:
        if art_words[i] not in G.nodes():
            G.add_node(art_words[i])
            G.add_edges_from([(art_words[i], art_words[i-1]), (art_words[i], art_words[i-2]), (art_words[i-1], art_words[i-2])])
            if art_words[i+1] not in G.nodes():
                G.add_node(art_words[i+1])
            elif art_words[i+2] not in G.nodes():
                G.add_node(art_words[i+2])
            G.add_edges_from([(art_words[i], art_words[i+1]), (art_words[i], art_words[i+2]), (art_words[i+1], art_words[i+2])])
        else:
            G.add_edges_from([(art_words[i], art_words[i-1]), (art_words[i], art_words[i-2]), (art_words[i-1], art_words[i-2])])
            if art_words[i+1] not in G.nodes():
                G.add_node(art_words[i+1])
            elif art_words[i+2] not in G.nodes():
                G.add_node(art_words[i+2])
            G.add_edges_from([(art_words[i], art_words[i+1]), (art_words[i], art_words[i+2]), (art_words[i+1], art_words[i+2])])
```

```python
nx.write_gexf(G, 'SovSport_graph.gexf')
```

```python
graph_radius = nx.radius(G)
print(graph_radius)
```

```python
graph_diameter = nx.diameter(G)
print(graph_diameter)
```

```python
nodes_nmb = G.number_of_nodes()
edges_nmb = G.number_of_edges()
print(nodes_nmb)
print(edges_nmb)
```

```python
graph_dens = nx.density(G)
print(graph_dens)
```

```python
graph_assortat = nx.degree_pearson_correlation_coefficient(G)
print(graph_assortat)
```

```python
node_imp = nx.degree_centrality(G)

for nodeid in sorted(node_imp, key=node_imp.get, reverse=True):
    print(nodeid, node_imp[nodeid])
```

```python
graph_clust = nx.average_clustering(G)
graph_trans = nx.transitivity(G)

print(graph_clust)
print(graph_trans)
```
